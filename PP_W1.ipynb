```
A = np.arange(1,10).reshape(3,3)
B = np.arange(2,11).reshape(3,3)
print("A",A)
print("B",B)

print("A @ B", A @ B)
print("A + B", np.add(A,B))

print("CUMSUM", np.cumsum(A))
print("SUM", A.sum(), "SUM BY ROWS", A.sum(1), "CUM BY CLMNS",A.sum(0))
```
