```
import requests
from bs4 import BeautifulSoup

data = requests.get('https://pogoda.interia.pl/prognoza-dlugoterminowa-bydgoszcz,cId,2942339')
# print(data.text)
html = BeautifulSoup(data.content, 'html.parser')
class_name = 'weather-currently-temp-strict'
temperature = html.find("div", attrs={'class' : class_name})
type(temperature)
print(str(temperature).replace('<div class="weather-currently-temp-strict">',"").replace("</div>",""))
```
